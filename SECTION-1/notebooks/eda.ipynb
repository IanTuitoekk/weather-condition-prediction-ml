```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# style for clean visual plots
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load cleaned dataset
df = pd.read_csv('../data/GlobalWeatherRepository_cleaned.csv')


df.head()
```

```python
# Check missing values and data types 
print(df.info())

# Summary statistics for numerical features
df.describe()
```

```python
# Strip whitespace and make everything title case to merge duplicates like "Partly cloudy" and "Partly Cloudy"
df['condition_text'] = df['condition_text'].str.strip().str.title()
```

```python
# Get the top 10 most common weather conditions
top_10_conditions = df['condition_text'].value_counts().head(10).index

# Filter dataset for only these top 10 categories
df_top10 = df[df['condition_text'].isin(top_10_conditions)]

plt.figure(figsize=(10, 5))
sns.countplot(
    data=df_top10, 
    y='condition_text', 
    order=top_10_conditions, 
    palette='viridis'
)
plt.title('Top 10 Most Frequent Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.tight_layout()
plt.savefig('../reports/distribution_weather_conditions.png', dpi=300)
plt.show()
```

```python
sample_numeric_cols = ['temperature_celsius', 'latitude', 'longitude', 'air_quality_PM2.5']

for col in sample_numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    
    # Histogram / KDE
    sns.histplot(df[col], kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Distribution of {col}')
    
    # Boxplot
    sns.boxplot(x=df[col], ax=axes[1], color='salmon')
    axes[1].set_title(f'Boxplot of {col}')
    
    plt.tight_layout()
    
    
    plt.savefig(f'../reports/boxplot_{col}.png', dpi=300)
    plt.show()
```

```python
# Select a clean, non-redundant subset of numerical columns
clean_subset = [
    'temperature_celsius', 'feels_like_celsius', 'humidity', 'cloud', 
    'pressure_mb', 'precip_mm', 'wind_kph', 'visibility_km', 
    'uv_index', 'air_quality_PM2.5', 'latitude'
]

# Calculate correlation on the clean subset
subset_corr = df[clean_subset].corr()

# Plot a much cleaner, readable heatmap with numbers included
plt.figure(figsize=(10, 8))
sns.heatmap(
    subset_corr, 
    annot=True,          # This puts the actual correlation numbers in the squares!
    cmap='coolwarm', 
    fmt=".2f", 
    linewidths=0.5, 
    center=0
)
plt.title('Cleaned Correlation Matrix (Key Features)')
plt.tight_layout()
plt.savefig('../reports/correlation_heatmap.png', dpi=300)
plt.show()
```
